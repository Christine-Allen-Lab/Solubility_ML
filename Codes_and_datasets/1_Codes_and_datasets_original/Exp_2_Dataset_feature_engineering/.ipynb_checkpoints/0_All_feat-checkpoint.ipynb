```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import lightgbm as lgb
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt

from sklearn.model_selection import GroupKFold, cross_val_predict
```

```python
file_path = "../Figure_2_Dataset_processing/Processed_dataset_20240115.csv"

data = pd.read_csv(file_path)
```

```python
train = data[data['Type'] == 'Train']
test = data[data['Type'] == 'Test']
lab = data[data['Type'] == 'Lab']
```

```python
cols_to_drop = ['Web of Science Index', 'Drug','Solvent_1', 'Solvent_2', 'Solubility (mol/mol)', 
                'DOI', 'Drug-solvent system', 'Solvent system','Solubility (g/100g)', 'LogS', 'Class',  'Type']
```

```python
X_train = train.drop(cols_to_drop, axis = 1)
X_test = test.drop(cols_to_drop, axis = 1)
X_lab = lab.drop(cols_to_drop, axis = 1)
```

```python
for col in X_train.columns:
    print(col)
```

```python
scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_lab_scaled = scaler.transform(X_lab)
```

```python
X = X_train_scaled
Y = train['LogS']
G = train['Drug-solvent system']

group_kfold = GroupKFold(n_splits=10)
```

```python
model = lgb.LGBMRegressor(random_state = 0, n_jobs = -1)
Y_pred = cross_val_predict(model, X, Y, groups=G, cv=group_kfold)


mae = mean_absolute_error(Y, Y_pred)
rmse = sqrt(mean_squared_error(Y, Y_pred))
r2  = r2_score(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
```

```python
X.shape
```

```python
places = 2
print("10 fold GKF CV:")
print("\n")
print("# of all features:     ", X.shape[1])
print("# of features used:    ", X.shape[1])
print("# of features removed: ", 0)
print("MAE:                   ", round(mae,places))
print("RMSE:                  ", round(rmse,places))
print("MSE:                   ", round(mse,places))
print("R2:                    ", round(r2,places))
```
