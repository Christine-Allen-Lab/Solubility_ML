```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
import ast
import shap
from matplotlib.lines import Line2D


from sklearn.model_selection import GroupKFold, cross_val_predict
import matplotlib.pyplot as plt
```

```python
file_path = "../Figure_2_Dataset_processing/Processed_dataset_20240125.csv"

data = pd.read_csv(file_path)
```

```python
data = data.sort_values('Type', ascending=False)
data = data.reset_index(drop = True)
data
```

```python
train = data[data['Type'] == 'Train']
test = data[data['Type'] == 'Test']
lab = data[data['Type'] == 'Lab']
```

```python
cols_to_drop = ['Web of Science Index', 'Drug', 'Solvent_1', 'Solvent_2', 'Solubility (mol/mol)', 'DOI', 'Drug-solvent system','Solvent system','Solubility (g/100g)', 'LogS', 'Class',  'Type']

```

```python
X_train = train.drop(cols_to_drop, axis = 1)
X_test = test.drop(cols_to_drop, axis = 1)
X_lab = lab.drop(cols_to_drop, axis = 1)
```

```python
for col in X_train.columns:
    print(col)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_lab_scaled = scaler.transform(X_lab)
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns = X_test.columns)
X_lab_scaled_df = pd.DataFrame(X_lab_scaled, columns = X_lab.columns)
```

```python
X_test_scaled_df
```

```python
# Function to perform feature importance reduction using Group K-Fold CV
def feature_importance_reduction(X, Y, G, n_splits):
    SHAP_summary = pd.DataFrame(columns=['Iteration', 'Feature to remove', 'Features Used', 'Num Features Used', 'MAE'])

    iteration = -1
    group_kfold = GroupKFold(n_splits=n_splits)

    while X.shape[1] > 1:
        iteration += 1

        model = lgb.LGBMRegressor(random_state=0, n_jobs=-1)
        Y_pred = cross_val_predict(model, X, Y, groups=G, cv=group_kfold)
        oob_mae = mean_absolute_error(Y, Y_pred)
        
        # Train the model on the full dataset
        model.fit(X, Y)

        # Calculate SHAP values
        explainer = shap.Explainer(model)
        shap_values = explainer(X)

        # Calculate the mean absolute SHAP values for each feature
        mean_abs_shap_values = np.abs(shap_values.values).mean(axis=0)
        
        # Identify the least important feature based on SHAP values
        least_important_feature = X.columns[np.argmin(mean_abs_shap_values)]

        # Prepare a new row for the summary DataFrame
        new_row = pd.DataFrame({
            'Iteration': iteration,
            'Feature to remove': least_important_feature,
            'Features Used': [list(X.columns)],
            'Num Features Used': X.shape[1],
            'MAE': oob_mae
        }, index=[0])

        SHAP_summary = pd.concat([SHAP_summary, new_row], ignore_index=True)

        print(f"Iteration: {iteration}, Feature to remove: {least_important_feature}, Num Features Used: {X.shape[1]}, MAE: {round(oob_mae,3)}")

        X = X.drop(columns=[least_important_feature])

    # Handling the last feature
    iteration += 1
    Y_pred = cross_val_predict(model, X, Y, groups=G, cv=group_kfold)
    oob_mae = mean_absolute_error(Y, Y_pred)
    new_row = pd.DataFrame({
        'Iteration': iteration,
        'Feature to remove': 'N/A',
        'Features Used': [list(X.columns)],
        'Num Features Used': X.shape[1],
        'MAE': oob_mae
    }, index=[0])
    SHAP_summary = pd.concat([SHAP_summary, new_row], ignore_index=True)
    print(f"Iteration: {iteration}, Feature to remove: 'N/A', Num Features Used: {X.shape[1]}, MAE: {round(oob_mae,3)}")

    return SHAP_summary
```

```python
SHAP_summary = feature_importance_reduction(X_train_scaled_df, train['LogS'], train['Drug-solvent system'], n_splits = 10)


```

```python
SHAP_summary.to_csv("Summary_and_dataset/SHAP_Summary.csv", index = False)
```

```python
def plot_accuracy(FI_summary, best_iteration, start_from):
    fontsize = 16  # Font size for all text elements
    
    red = '#DD706E'
    grey = '#515265'
    yellow = '#FAAF3A'
    blue = '#3A93C2'

    best_iteration = FI_summary[FI_summary['Iteration'] == best_iteration]
    # Plotting
    fig, ax = plt.subplots(figsize=(12, 6))
    FI_summary = FI_summary[FI_summary['Num Features Used']>= start_from]
    ax.plot(FI_summary['Num Features Used'], FI_summary['MAE'], marker='o', markersize=8, markeredgewidth=0.25, markeredgecolor='black', 
         linestyle='-', color=red,)
    
    #ax.scatter(FI_summary['Num Features Used'], FI_summary['MAE'], color=red)

    # Highlight the best iteration
    best_num_features = best_iteration['Num Features Used'].values
    best_accuracy = best_iteration['MAE'].values
    ax.scatter(best_num_features, best_accuracy, color=grey, zorder=5, s=70, edgecolor = 'black', linewidths = 0.5)

    # Annotate the last scatter dot
    ax.annotate(f'({best_num_features[0]}, {best_accuracy[0]:.3f})', 
                xy=(best_num_features[0], best_accuracy[0]), 
                xytext=(best_num_features[0]+10, best_accuracy[0]+0.03),
                fontsize=fontsize,
                arrowprops=dict(facecolor='black', arrowstyle='->'))

    # Create custom legend
    legend_elements = [Line2D([0], [0], marker='o', color='w', label='Best feature set',
                              markerfacecolor=grey, markersize=10, markeredgecolor='black', markeredgewidth=0.5),
                       Line2D([0], [0], marker='o', color='w', label=f'Number of features: {best_num_features[0]}',
                              markerfacecolor='none', markersize=0),
                       Line2D([0], [0], marker='o', color='w', label=f'MAE: {best_accuracy[0]:.3f}',
                              markerfacecolor='none', markersize=0)]
    
    legend = ax.legend(handles=legend_elements, title="SHAP-based feature screening",
              loc='upper right', framealpha=0, edgecolor='none', fontsize=fontsize-2)
    legend.get_title().set_fontsize(fontsize-2)

    # Set title and labels
    ax.set_title('', fontsize=fontsize)
    ax.set_xlabel('Number of features', fontsize=fontsize)
    ax.set_ylabel('Cross-validation performance (MAE)', fontsize=fontsize-2)
    ax.set_ylim(0.34, 0.55)

    # Set tick labels fontsize
    ax.tick_params(axis='both', which='major', labelsize=fontsize)

    # Remove grid
    ax.grid(False)
    ax.set_facecolor('none')
    fig.patch.set_alpha(0.0)
    ax.annotate('c)', xy=(0.0, 1.06), xycoords="axes fraction", va="top", ha="left", fontsize=fontsize)
    
    fig.canvas.print_figure('Figure_3c_SHAP.png', dpi=600)
    plt.show()
```

```python
def dataset_generator(X_train_array, X_test_array, X_lab_array, data, all_feat, feat_to_keep, col_to_add):
    
    X_train = pd.DataFrame(X_train_array, columns = all_feat)
    X_test = pd.DataFrame(X_test_array, columns = all_feat)
    X_lab = pd.DataFrame(X_lab_array, columns = all_feat)
    
    X_train = X_train[feat_to_keep]
    X_test = X_test[feat_to_keep]
    X_lab = X_lab[feat_to_keep]
    
    X_train['Type'] = 'Train'
    X_test['Type'] = 'Test'
    X_lab['Type'] = 'Lab'
    
    combined_df = pd.concat([X_train, X_test, X_lab], ignore_index=True)
    
    for col in col_to_add:
        combined_df[col] = data[col]
        
    train_subset = combined_df[combined_df['Type'] == 'Train']
    test_subset = combined_df[combined_df['Type'] == 'Test']
    

    overlapping_groups = set(train_subset['Drug-solvent system']).intersection(set(test_subset['Drug-solvent system']))
    
    if (combined_df['Type'].equals(data['Type'])):
        print('First check:  passed')     
    else:
        print('First check:  failed')
        
    if (combined_df['LogS'].equals(data['LogS'])):
        print('Second check: passed')     
    else:
        print('Second check: failed')
        
    if (combined_df['Drug'].equals(data['Drug'])):
        print('Third check:  passed')     
    else:
        print('Third check:  failed')
    
    print("Number of overlapping train/test groups: ", len(overlapping_groups))
    
    return combined_df
```

```python
combined_df = dataset_generator(X_train_array = X_train_scaled, X_test_array = X_test_scaled, 
                                X_lab_array = X_lab_scaled,
                                data = data, all_feat = X_train.columns, 
                                feat_to_keep = feat_to_keep, col_to_add = col_to_add)

```

```python
combined_df.to_csv('Summary_and_dataset/SHAP_dataset.csv', index = False)
```

```python
train_subset = combined_df[combined_df['Type'] == 'Train']
```

```python
X = train_subset[feat_to_keep]
Y = train_subset['LogS']
G = train_subset['Drug-solvent system']

group_kfold = GroupKFold(n_splits=10)
```

```python
model = lgb.LGBMRegressor(random_state = 0, n_jobs = -1)
Y_pred = cross_val_predict(model, X, Y, groups=G, cv=group_kfold)


mae = mean_absolute_error(Y, Y_pred)
medae = np.median(np.abs(Y - Y_pred))
rmse = sqrt(mean_squared_error(Y, Y_pred))
mse = mean_squared_error(Y, Y_pred)
```

```python
places = 2
print("10 fold GKF CV:")
print("\n")
print("# of all features:     ", len(X_train.columns))
print("# of features used:    ", len(feat_to_keep))
print("# of features removed: ", len(X_train.columns) - len(feat_to_keep))
print("MAE:  ", round(mae,places))
print("MedAE:", round(medae,places))
print("RMSE: ", round(rmse,places))
print("MSE:  ", round(mse,places))
```
